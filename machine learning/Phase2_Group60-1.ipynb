```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format = 'retina'
plt.style.use("ggplot")

pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv("Phase2_Group60.csv")
data.head()
```

```python
corr_data = data.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_data, cmap="Reds", annot=True)
```

```python
data.drop(['euribor3m', 'nr.employed', 'duration', 'emp.var.rate', 'cons.price.idx', 'day_of_week', 'month'], axis=1, inplace=True)
```

# Splitting the Data and Target variables

```python
# Separate features (X) and target variable (y)
X = data.drop(columns=['subscribed'])  # Assuming 'subscribed' is the target variable
y = data['subscribed']
y.value_counts()
```

# One Hot Encoding

```python
# Converting Binary type columns to numerical

binary_cols = ['housing', 'loan', 'subscribed', 'default']
dummy_list = []

for i in binary_cols:
    for text in data[i].values:
        if text == 'yes':
            dummy_list.append(1)
        else:
            dummy_list.append(0)
    data[i] = dummy_list
    dummy_list = []
    
```

```python
# Converting categorical columns to numerical 
categoricalColumns = X.columns[X.dtypes==object].tolist()

for col in categoricalColumns:
    print('Unique values for ' + col)
    print(X[col].unique())
    print('')

for col in categoricalColumns:
    n = len(X[col].unique())
    if (n == 2):
        X[col] = pd.get_dummies(X[col], drop_first=True)

X = pd.get_dummies(X)

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

```

```python
X.sample(10, random_state=40)
```

# Scaling

```python
from sklearn import preprocessing

X_data = X.copy()
X_scaler = preprocessing.MinMaxScaler()
X_scaler.fit(X)
X = X_scaler.fit_transform(X)
```

```python
pd.DataFrame(X, columns=X_data.columns).sample(5, random_state=999)
```

# Predictive Modelling:<a name = "predictive" ></a>

The process of selecting features is an essential part of predictive modeling since it aids in determining which features are most relevant to the prediction of the target variable. ## Performance for all features

Initially, we assessed the performance of a decision tree classifier using all features in the dataset. The model was evaluated using repeated stratified K-fold cross-validation with ROC-AUC as the scoring metric. The results indicated a mean ROC-AUC score was 0.619.

```python
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(random_state=999)
cv_meth = RepeatedStratifiedKFold(n_splits = 5,
                                  n_repeats=3,
                                  random_state=999)
scoring_metric = 'roc_auc'
cv_results = cross_val_score(estimator=clf,
                             X=X,
                             y=y,
                             cv=cv_meth,
                             scoring=scoring_metric)
```

```python
cv_results
```

```python
cv_results.mean().round(3)
```

# F Score

We employed the F-score method to select the top 10 features based on their importance. The selected features were plotted, and the model performance was reassessed using only these features. The mean ROC-AUC score with F-score selected features was 0.721.

```python
from sklearn import feature_selection

num_features=10


fs_fit_fscore = feature_selection.SelectKBest(feature_selection.f_classif, k=num_features)
fs_fit_fscore.fit_transform(X, y)
fs_indices_fscore = np.argsort(np.nan_to_num(fs_fit_fscore.scores_))[::-1][0:num_features]
fs_indices_fscore
```

```python
best_fscores = X_data.columns[fs_indices_fscore].values
best_fscores
```

```python
feature_imp = fs_fit_fscore.scores_[fs_indices_fscore]
feature_imp
```

```python
def best_feat_plot(best_feat, scores, method):
    plt.barh(best_feat, scores)
    plt.title(method + ' Feature importance for top 10')
    plt.xlabel("Importance")
    plt.ylabel("Features")
    plt.show()

best_feat_plot(best_fscores, feature_imp, 'F-Score')
```

```python
fscore_results = cross_val_score(estimator=clf,
                             X=X[:, fs_indices_fscore],
                             y=y,
                             cv=cv_meth,
                             scoring=scoring_metric)

fscore_results.mean().round(3)
```

# Model Fitting and Tuning <a name = "fit" ></a>

We evaluated several machine learning models to identify the best-performing model for our dataset. The models included:

1. Random Forest
2. Gradient Boosting

Each model was trained and evaluated using the selected features from the Random Forest importance method. The evaluation metrics included accuracy, accuracy_mean, recall, F1-score, ROC-AUC and standard deviation.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=999, stratify=y)


print(X_test.shape)
print(X_train.shape)
```

```python
# Selecting the 10 best features

X_train_fs = X_train[:, fs_indices_fscore]
X_test_fs = X_test[:, fs_indices_fscore]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, ConfusionMatrixDisplay, precision_score, recall_score, f1_score, classification_report, roc_curve, auc, precision_recall_curve, average_precision_score
from sklearn.model_selection import cross_val_score
from scikitplot.metrics import plot_roc_curve, plot_precision_recall_curve
```

```python
models = []
models.append(['Random Forest', RandomForestClassifier(random_state=0)])
models.append(['Gradient Boosting', GradientBoostingClassifier(random_state=0)])
models.append(['SVM', SVC(random_state=0)])
models.append(['GaussianNB', GaussianNB()])
models.append(['KNeigbors', KNeighborsClassifier()])
models.append(['Decision Tree', DecisionTreeClassifier(random_state=0)])
```

```python
list_1 = []

for i in range(len(models)):
    list_2= []
    model = models[i][1]
    model.fit(X_train_fs, y_train)
    y_pred = model.predict(X_test_fs)
    cm = confusion_matrix(y_test, y_pred) 
    accuracies = cross_val_score(estimator = model, X = X_train_fs, y = y_train, cv = 10)   #K-Fold Validation
    roc = roc_auc_score(y_test, y_pred)  
    precision = precision_score(y_test, y_pred)  
    recall = recall_score(y_test, y_pred)  
    f1 = f1_score(y_test, y_pred)  
    print(models[i][0],':')
    print(cm)
    print('Accuracy Score: ', accuracy_score(y_test, y_pred))
    print('')
    print('K-Fold Validation Accuracy: {:.2f}%'.format(accuracies.mean()*100))
    print('')
    print('ROC AUC score: {:.2f}%'.format(roc))
    print('')
    print('Standard Deviation: {:.2f}%'.format(accuracies.std()*100))
    print('')
    print('Recall: {:.2f}%'.format(recall))
    print('')
    print('F1: {:.2f}%'.format(f1))
    print('-----------------------------')
    print('')
    list_2.append(models[i][0])
    list_2.append((accuracy_score(y_test, y_pred))*100)
    list_2.append(accuracies.mean()*100)
    list_2.append(accuracies.std()*100)
    list_2.append(roc)
    list_2.append(recall)
    list_2.append(precision)
    list_2.append(f1)
    list_1.append(list_2)
```

```python
df= pd.DataFrame(list_1, columns = ['Model', 'Accuracy', 'K-Fold Mean Accuracy', 'Std. Deviation', 'ROC AUC', 'Precision', 'Recall', 'F1'])
df.sort_values(by=['Accuracy', 'K-Fold Mean Accuracy'], inplace=True, ascending=False)
df
```

# Critique & Limitations <a name="limit"></a>

### Strengths

1. Comprehensive Feature Selection: We used both Random Forest significance and the F-score technique in our feature selection process. We were able to determine which characteristics were most important, which enhanced the interpretability and performance of the model.

2. Diverse Model Evaluation: We evaluated a diverse set of machine learning models, including traditional algorithms. This ensured a robust comparison and helped us understand which models are best suited for our specific problem.

3. Extensive Hyperparameter Tuning: We maximized each model's performance through hyperparameter tuning utilizing grid search and random search strategies. Our forecasts were more reliable and accurate as a result of this detailed tuning procedure.

4. Cross-Validation and Statistical Testing: A strict framework for evaluating the model was made possible by the use of paired t-tests and cross-validation. Our results were guaranteed to be statistically significant and not the product of random chance thanks to this technique.

### Limitations

1. Imbalanced Dataset: The imbalance in the dataset was one of the main drawbacks. The target variable, which indicated whether or not a client had a term deposit, was skewed, with non-subscribers making up the bulk of the cases. Models that are biased in favor of the majority class may result from this imbalance. Future work might benefit from methods particularly developed to handle unbalanced datasets, including class-weight modifications or SMOTE (Synthetic Minority Over-sampling Technique). This is even if assessment techniques like ROC-AUC are still being used.

2. Computational Intensity:Some models required a lot of computing power because of the large dimensionality of the data, especially the K-Nearest Neighbors (KNN). Because KNN needs to calculate the distance between each pair of instances, it is known to be sluggish when dealing with huge datasets.

3. Overfitting Risk: Extensive hyperparameter tuning always carries the risk of overfitting, in which the model excels on training data but is unable to generalize to new data. While cross-validation reduces this risk, in order to guarantee the model's generalizability, it is essential to validate it on an entirely separate test set.

4. Model Interpretability: Certain models, like Random Forests and Decision Trees, may be somewhat interpreted, while other models, like Support Vector Machines, are frequently referred to as "black boxes." When attempting to explain the model's judgments to stakeholders or regulatory agencies, this lack of transparency may be a hindrance.

5. Feature Engineering: Even with our thorough feature selection, more feature engineering can enhance the performance of the model even more. For example, generating polynomial features or interaction terms might help identify more intricate correlations in the data. Nevertheless, this might exacerbate computing difficulties by adding to the models' dimensions and complexity.

6. Temporal Data Issues: Although the dataset lacks explicit time-series data, temporal connections may be introduced by some variables (such as the month of contact). Ignoring them might result in models that are unable to accurately represent changes over time or seasonal trends. More study might investigate openly adding temporal factors.

7. Economic Indicators: A number of economic variables that may have changed dramatically over time were included in the dataset. The dynamic character of these indicators might not be well captured by static models. For deeper insights, a more dynamic modeling technique that could incorporate time-series analysis could be helpful.

8. Scalability: Scalability would need to be taken into account when implementing these models in a real-world setting. The KNN model in particular may require high-performance computer resources because to its computational intensity. Furthermore, for real-time applications, the delay in predicting using models such as KNN must be addressed.
